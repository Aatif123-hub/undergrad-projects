# 1. Preprocessing: Handling Missing Data and Preparing the Dataset

```python
import pandas as pd
import numpy as np

weather_data = pd.read_csv('weather.csv')

print("Missing values in each column:")
print(weather_data.isnull().sum())

weather_data.fillna(weather_data.mean(), inplace=True)

X = weather_data[['humidity', 'wind_speed', 'meanpressure']]
y = weather_data['meantemp']
```

# 2. Model Training: Training a Decision Tree Regression Model and Calculating Performance Metrics


```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```

# 3. Visualizations: Plotting the Actual vs Predicted Temperature Values

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label='Actual Temperature', color='blue')
plt.plot(y_pred, label='Predicted Temperature', color='red', linestyle='--')
plt.xlabel('Sample Index')
plt.ylabel('Temperature')
plt.title('Actual vs Predicted Temperature')
plt.legend()
plt.show()
```
